```python
import pandas as pd
import numpy as np
```

```python
diabetes = pd.read_csv('diabetes.csv')

diabetes.head(10)
```

```python
diabetes.shape
```

```python
diabetes.info()
```

```python
diabetes.describe().transpose()

#Here we can see that for Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin and BMI min value is 0. 
#For Pregnancies it is possible but for the rest it is impossible. 
#So that means these values are missing values
```

```python
diabetes['Glucose'].replace(0, np.nan, inplace= True)
diabetes['BloodPressure'].replace(0, np.nan, inplace= True)
diabetes['SkinThickness'].replace(0, np.nan, inplace= True)
diabetes['Insulin'].replace(0, np.nan, inplace= True)
diabetes['BMI'].replace(0, np.nan, inplace= True)

#Replace 0 values with NaN
```

```python
diabetes.isnull().sum()
#Sum the null values in every column

#We can see that Insulin and SkinThickness have a lot of missing values
#These are the two biggest columns to fix
```

```python
arr = diabetes['SkinThickness'].values.reshape(-1, 1)

arr.shape

#Let's work on SkinThickness column first. Reshape in the form of a 2D array 
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

imp.fit(diabetes['SkinThickness'].values.reshape(-1, 1))

diabetes['SkinThickness'] = imp.transform(diabetes['SkinThickness'].values.reshape(-1, 1))

#In order to fill in missing values using inference from the current data we have available, use
#SimpleImputer estimator from SciKit-Learn 
#SimpleImputer offers a number of basic strategies for imputing or inferring missing values 
#You can fill in using a constant, the mean of the existing data, median, mode, etc.

#In the case of the Skin Thickness column, the stategy that will be used to fill in missing values will be the mode/most frequently seen value
```

```python
diabetes['SkinThickness'].describe()

#When you use imputation techniques to fill in missing values, the statistical properties of your data will change 
#You have to be aware of this because imputation techniques will change the characteristics of your data and you have to be ok with this change based on your dataset and use case
#We don't have much data to work with, so we don't want to drop records with missing values and so that's why imputation works best
```

```python
diabetes.isnull().sum()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='median')

imp.fit(diabetes['Glucose'].values.reshape(-1, 1))

diabetes['Glucose'] = imp.transform(diabetes['Glucose'].values.reshape(-1, 1))

#Here Strategy = 'median' which means it will replace missing values using the median in the column Glucose.
```

```python
diabetes['Glucose'].describe()

#You can see using this strategy that the mean didn't change that much and neither did the std
#So this is a great strategy 
```

```python
diabetes.isnull().sum()
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')

imp.fit(diabetes['BloodPressure'].values.reshape(-1, 1)) #fit being used on a series/column/record

diabetes['BloodPressure'] = imp.transform(diabetes['BloodPressure'].values.reshape(-1, 1)) #transform being used on the data frame/array

#Here Strategy = 'mean' which means it will replace missing values using the mean in the column.
```

```python
diabetes['BloodPressure'].describe()

#The mean didn't change all that much since we used the mean strategy to fill in missing values after all
#But the std changed a lot, so this isn't the best strategy either -- because we imputed using the mean value
#mode values are centered at the mean
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value=32)

imp.fit(diabetes['BMI'].values.reshape(-1, 1))

diabetes['BMI'] = imp.transform(diabetes['BMI'].values.reshape(-1, 1))

#Now use constant strategy 
```

```python
diabetes['BMI'].describe()
#mean stayed same (because 32 is the mean anyways), but std changed quite a bit 
```
